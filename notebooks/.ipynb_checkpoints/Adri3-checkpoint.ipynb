```python
df2=df2.reset_index()
df2.head()
```

```python
df2.Item.unique()
```

```python
df2.drop(df2[df2['Item'] != 'All sectors without LULUCF'].index, inplace = True)
df2.info
```

```python
df2.Element.unique()
```

```python
CO2_df = df2.loc[df2["Element"] == 'Emissions (CO2)']
```

```python
N20_df = df2.loc[df2["Element"] == 'Emissions (N2O)']
```

```python
CH4_df = df2.loc[df2["Element"] == 'Emissions (CH4)']
```

```python
CO2_df = CO2_df[["Area","2018"]].copy()
CO2_df
```

```python
CO2_df.to_csv("../data/CO2_DF.csv")
```

```python
N20_df = N20_df[["Area","2018"]].copy()
N20_df
```

```python
N20_df.to_csv("../data/N20_df.csv")
```

```python
CH4_df = CH4_df[["Area","2018"]].copy()
CH4_df
```

```python
CH4_df.to_csv("../data/CH4_df.csv")
```

```python
EM_df = CO2_df.merge(N20_df,on='Area').merge(CH4_df,on='Area')
print(EM_df)
```

```python
EM_df = EM_df.rename({"2018_x":"CO2","2018_y":"N20","2018":"CH4"}, axis=1)
```

```python
EM_df = EM_df.set_index("Area")
```

```python
norm_EM_df=(EM_df-EM_df.mean())/EM_df.std()
norm_EM_df
```

```python
norm_EM_df['Total_emissions'] = norm_EM_df["CO2"] + norm_EM_df["N20"] + norm_EM_df["CH4"]
norm_EM_df
```

```python
norm_EM_df.to_csv("../data/TOT_EM_DF.csv")
```
