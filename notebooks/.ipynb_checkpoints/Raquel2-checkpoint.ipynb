```python
import pandas as pd
```

```python
car = pd.read_csv('../data/car_clean.csv')
co = pd.read_csv('../data/co2_clean.csv')
oce = pd.read_csv('../data/oce_clean.csv')
ter = pd.read_csv('../data/ter_clean.csv')
wat = pd.read_csv('../data/wat_clean.csv')
```

```python
co.drop(['Unnamed: 0'], axis =1, inplace=True, errors = 'ignore')
co.rename( columns ={'2020': 'co20'}, inplace = True)
co
```

```python
oce.drop(['Unnamed: 0'], axis =1, inplace=True, errors = 'ignore')
oce.rename( columns ={'2020': 'oce20'}, inplace = True)
oce
```

```python
all = pd.merge(co, oce, on='Country')
```

```python
all
```

```python
ter.drop(['Unnamed: 0'], axis =1, inplace=True, errors = 'ignore')
ter.rename( columns ={'2020': 'ter20'}, inplace = True)
ter
```

```python
all = pd.merge(all, ter, on='Country')
```

```python
all
```

```python
wat.drop(['Unnamed: 0'], axis =1, inplace=True, errors = 'ignore')
wat.rename( columns ={'2020': 'wat20'}, inplace = True)
wat
```

```python
all = pd.merge(all, wat, on='Country')
```

```python
all
```

```python
all_sorted = all.sort_values("co20", ascending=False).reset_index(drop=True)
```

```python
all_sorted.head(20)
```

```python
all_sorted.to_csv('../data/all_sorted.csv')
```
